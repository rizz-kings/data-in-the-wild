```python
from datetime import datetime
from time import strftime

today = datetime.now().strftime("%Y-%m-%d")

total_pages = tmdb.Discover().movie(with_keywords='180547', without_keywords='263548')['total_pages']

movie_list = []
movie_ids = []

for page in range(1, total_pages+1):
    current_page = tmdb.Discover().movie(with_keywords='180547', without_keywords='263548', page=page)
    for movie in current_page['results']:
        if "release_date" in movie and (movie["release_date"] < today and movie['release_date'] != ''):
            movie_name = movie['title']
            print(movie_name)
            movie_list.append(movie_name)
            movie_ids.append(movie['id'])
```

```python
movie_list_filtered = [i for i in movie_list if not ('Marvel Studios' in i)]
movie_ids_filtered = [i for i in movie_ids if movie_list[movie_ids.index(i)] in movie_list_filtered]
for i in range(len(movie_list_filtered)):
    movie_list_filtered[i] = movie_list_filtered[i].lower()
    movie_list_filtered[i] = movie_list_filtered[i].replace(":", "")
    movie_list_filtered[i] = movie_list_filtered[i].replace(" ", "-")
len(movie_ids_filtered)
```

```python
from json import loads, dumps
json_object = dumps(movie_list_filtered, indent=4)
```

```python
with open("mcu_list.json", "w") as outfile:
    outfile.write(json_object)
```

```python
movie_series = {
    "movie_name": [],
    "imdb_id": [],
    "cast": [],
    "Direction": [],
    "movie_budget": [],
    "movie_revenue": []
    
}
for id in movie_ids_filtered:
    current_movie = tmdb.Movies(id)
    current_movie.info()
    movie_series["movie_name"].append(current_movie.original_title)
    movie_series["imdb_id"].append(current_movie.imdb_id)
    movie_series["movie_budget"].append(current_movie.budget)
    movie_series["movie_revenue"].append(current_movie.revenue)
    movie_cast = []
    for cast in current_movie.credits()["cast"]:
        if cast["popularity"] > 10:
            movie_cast.append((cast["name"], cast["character"]))
    movie_series["cast"].append(movie_cast)
```
