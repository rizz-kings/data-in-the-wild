```python
from datetime import datetime
from time import strftime

today = datetime.now().strftime("%Y-%m-%d")

total_pages = tmdb.Discover().movie(with_keywords=['180547'], without_keywords='263548')['total_pages']

movie_list = []
movie_ids = []

for page in range(1, total_pages+1):
    current_page = tmdb.Discover().movie(with_keywords=['180547'], without_keywords='263548', page=page)
    for movie in current_page['results']:
        if "release_date" in movie and (movie["release_date"] < today and movie['release_date'] != ''):
            movie_name = movie['title']
            if movie_name.find("Marvel Studios"):
                print(movie_name)
                movie_list.append(movie_name)
                movie_ids.append(movie['id'])
```

```python
print(len(movie_list))
```

```python
movie_series = {
    "movie_name": [],
    "imdb_id": [],
    "release_date": [],
    "cast": [],
    "direction": [],
    "production_companies": [],
    "movie_budget": [],
    "movie_revenue": []
    
}
for id in movie_ids:
    current_movie = tmdb.Movies(id)
    current_movie.info()
    movie_series["movie_name"].append(current_movie.title)
    movie_series["imdb_id"].append(current_movie.imdb_id)
    movie_series["release_date"].append(current_movie.release_date)
    movie_series["movie_budget"].append(current_movie.budget)
    movie_series["movie_revenue"].append(current_movie.revenue)
    movie_cast = {}
    directors = []
    production_companies = []
    for cast in current_movie.credits()["cast"]:
        if 'character' in cast and cast["popularity"] > 5:
            movie_cast[cast["name"]] = cast["character"]
    for crew in current_movie.credits()["crew"]:
        if crew['job'] == "Director":
            directors.append(crew["name"])
    for company in current_movie.production_companies:
        production_companies.append(company["name"])
        
    movie_series["cast"].append(movie_cast)
    movie_series["direction"].append(directors)
    movie_series["production_companies"].append(production_companies)
```

```python
len(movie_series["direction"])
```

```python
import pandas as pd



df = pd.DataFrame(movie_series)

df = df[df["movie_budget"] != 0]
df = df[df["movie_revenue"] != 0]
df = df[df["cast"] != {}]

df
```

```python
df.to_csv("../data/processed/movie_stats/movie_stats.csv")
```
