```python
from datetime import datetime
from time import strftime

today = datetime.now().strftime("%Y-%m-%d")

total_pages = tmdb.Discover().movie(with_keywords='180547', without_keywords='263548')['total_pages']

movie_list = []

for page in range(1, total_pages+1):
    current_page = tmdb.Discover().movie(with_keywords='180547', without_keywords='263548', page=page)
    for movie in current_page['results']:
        if "release_date" in movie and movie["release_date"] < today:
            movie_name = movie['title']
            print(movie_name)
            movie_list.append(movie_name)
```

```python
movie_list_filtered = [i for i in movie_list if not ('Marvel Studios' in i)]
for i in range(len(movie_list_filtered)):
    movie_list_filtered[i] = movie_list_filtered[i].lower()
    movie_list_filtered[i] = movie_list_filtered[i].replace(":", "")
    movie_list_filtered[i] = movie_list_filtered[i].replace(" ", "-")
```

```python
from json import loads, dumps
json_object = dumps(movie_list_filtered, indent=4)
```
